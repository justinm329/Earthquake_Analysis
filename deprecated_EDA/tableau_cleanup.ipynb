### Meausring the Mean & StDev

```python
mean_array = []
stdev_array = []
for i in graph_df.columns:
    mean_array.append(np.round(graph_df[i].mean()))
    stdev_array.append(np.round(graph_df[i].std()))

mean_array = np.array(mean_array)
stdev_array = np.array(stdev_array)
```

```python
from scipy.stats import weibull_min
```

```python
shape, loc, scale = weibull_min.fit(graph_df[0])
```

```python
shape, loc, scale
```

```python
random_numbers = weibull_min.rvs(shape, loc, scale, size=600)
```

```python
# Plot the histogram of the original data
plt.hist(graph_df[0], density=True, bins=10, alpha=0.5, label='Data')

# Generate x-values for the PDF plot
x = np.linspace(0, np.max(graph_df[0]), 100)

# Plot the PDF of the fitted Weibull distribution
plt.plot(x, weibull_min.pdf(x, shape, loc, scale), 'r-', label='Fitted Weibull PDF')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```
